```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/ratimayy/tools/main/directory.csv')
df.head()
```

image.png

```python
import pandas as pd
import plotly.graph_objects as go

# Prepare data
Country = df['Country']
number = df['Store Number']

# Create a bar chart with lighter shade of red bars
fig = go.Figure(data=go.Bar(x=Country, y=number, marker=dict(color='lightcoral')))

# Define layout with white background
fig.update_layout(
    title='Bar Chart',
    xaxis_title='Country',
    yaxis_title='Number of Stores',
    plot_bgcolor='white'  # Set background color to white
)

# Display the plot
fig.show()
```

```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df2 = pd.read_csv('https://raw.githubusercontent.com/ratimayy/tools/main/starbuck2.csv')
df2.head()
```

```python
#ตย map location
import pandas as pd
us_cities = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/us-cities-top-1k.csv")

import plotly.express as px

fig = px.scatter_mapbox(us_cities, lat="lat", lon="lon", hover_name="City", hover_data=["State", "Population"],
                        color_discrete_sequence=["fuchsia"], zoom=3, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Filter the DataFrame for rows where the value in the "City" column is "Chaing Mai"
filtered_df = df2[df2['City'] == 'Chaing Mai']

# Create scatter map plot using Plotly Express
fig = px.scatter_mapbox(filtered_df, lat="Latitude", lon="Longitude", hover_name="City", zoom=3, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
